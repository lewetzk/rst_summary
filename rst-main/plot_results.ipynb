```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("classification_result.txt", delimiter="\t",  header=None, usecols=[1,2, 3, 4, 5,6])
data
```

```python
types = ["random", "baseline", "snh", "gold", "neural25", "neural50"]
```

```python
ddict = {
    1: dict(),
    2: dict(),
    3: dict(),
    4: dict(),
    5: dict(),
    6: dict(),
}
```

```python
for i in range(6):
    for t in types:
        ddict[i+1][t] = (data[i+1] == t).sum()
```

```python
ddict
```

```python
to_plot = pd.DataFrame(ddict)
```

```python
to_plot.T.plot.line(figsize=(20, 15), linewidth=8)
plt.legend(types, ncol=3, loc='upper right')
plt.xlabel('ORDER 1(WORST) - 6 (BEST)', fontsize=18)
plt.ylabel('FREQUENCY', fontsize=16)
plt.xticks(fontsize=20, rotation=0)
```
